```javascript
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```javascript
img = cv2.imread("image.jpg")   # Put image.jpg in same folder

if img is None:
    print("❌ image.jpg not found!")
else:
    print("✅ Image loaded successfully!")
```

```javascript
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print("✅ BGR to RGB conversion done")
print("RGB Shape:", rgb.shape)

plt.imshow(rgb)
plt.title("Original RGB Image")
plt.axis("off")
plt.show()
```

```javascript
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("✅ Grayscale conversion done")
print("Gray Shape:", gray.shape)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```javascript
print("\n--- Image Details ---")
print("Shape :", img.shape)
print("Size  :", img.size)
print("Type  :", img.dtype)

rows, cols = gray.shape
print("Rows  :", rows)
print("Cols  :", cols)
```

```javascript
resized = cv2.resize(rgb, (300, 300))

print("✅ Image resized")
print("New Shape:", resized.shape)

plt.imshow(resized)
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```javascript
flip_horizontal = cv2.flip(rgb, 1)

print("✅ Horizontal flip done")

plt.imshow(flip_horizontal)
plt.title("Flip Horizontal")
plt.axis("off")
plt.show()
```

```javascript
flip_vertical = cv2.flip(rgb, 0)

print("✅ Vertical flip done")

plt.imshow(flip_vertical)
plt.title("Flip Vertical")
plt.axis("off")
plt.show()
```

```javascript
# alpha = contrast, beta = brightness
bright_contrast = cv2.convertScaleAbs(rgb, alpha=1.5, beta=30)

print("✅ Brightness + Contrast applied")
print("alpha (contrast) = 1.5")
print("beta (brightness) = 30")

plt.imshow(bright_contrast)
plt.title("Brightness + Contrast")
plt.axis("off")
plt.show()
```

```javascript
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

sharpened = cv2.filter2D(rgb, -1, sharpen_kernel)

print("✅ Sharpening done")
print("Kernel used:\n", sharpen_kernel)

plt.imshow(sharpened)
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```

```javascript
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian = np.uint8(np.absolute(laplacian))

print("✅ Laplacian operation done")

plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian")
plt.axis("off")
plt.show()
```

```javascript
# Move image right by 80 pixels and down by 50 pixels
translate_matrix = np.float32([[1, 0, 80],
                               [0, 1, 50]])

translated = cv2.warpAffine(rgb, translate_matrix, (cols, rows))

print("✅ Translation done")
print("Translation Matrix:\n", translate_matrix)

plt.imshow(translated)
plt.title("Translated Image")
plt.axis("off")
plt.show()
```

```javascript
# Rotate around center by 45 degrees
rotate_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)

rotated = cv2.warpAffine(rgb, rotate_matrix, (cols, rows))

print("✅ Rotation done")
print("Rotation Matrix:\n", rotate_matrix)

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```javascript
negative = 255 - rgb

print("✅ Negative transformation done")

plt.imshow(negative)
plt.title("Negative Transformation")
plt.axis("off")
plt.show()
```

```javascript
# Convert to float first
log_image = np.log1p(rgb.astype(np.float32))
log_image = cv2.normalize(log_image, None, 0, 255, cv2.NORM_MINMAX)
log_image = np.uint8(log_image)

print("✅ Log transformation done")

plt.imshow(log_image)
plt.title("Log Transformation")
plt.axis("off")
plt.show()
```

```javascript
gamma = 0.5   # <1 = brighter, >1 = darker
gamma_image = np.array(255 * (rgb / 255) ** gamma, dtype='uint8')

print("✅ Gamma transformation done")
print("Gamma value =", gamma)

plt.imshow(gamma_image)
plt.title("Gamma Transformation")
plt.axis("off")
plt.show()
```

```javascript
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

print("✅ Binary image created for morphology")

plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```javascript
kernel = np.ones((5, 5), np.uint8)

print("✅ Kernel created")
print("Kernel:\n", kernel)
```

```javascript
erosion = cv2.erode(binary, kernel, iterations=1)

print("✅ Erosion done")

plt.imshow(erosion, cmap='gray')
plt.title("Erosion")
plt.axis("off")
plt.show()
```

```javascript
dilation = cv2.dilate(binary, kernel, iterations=1)

print("✅ Dilation done")

plt.imshow(dilation, cmap='gray')
plt.title("Dilation")
plt.axis("off")
plt.show()
```

```javascript
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

print("✅ Opening done")
print("Opening = Erosion + Dilation")

plt.imshow(opening, cmap='gray')
plt.title("Opening")
plt.axis("off")
plt.show()
```

```javascript
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

print("✅ Closing done")
print("Closing = Dilation + Erosion")

plt.imshow(closing, cmap='gray')
plt.title("Closing")
plt.axis("off")
plt.show()
```

```javascript
gradient = cv2.morphologyEx(binary, cv2.MORPH_GRADIENT, kernel)

print("✅ Morphological Gradient done")
print("Gradient = Dilation - Erosion")

plt.imshow(gradient, cmap='gray')
plt.title("Morphological Gradient")
plt.axis("off")
plt.show()
```
